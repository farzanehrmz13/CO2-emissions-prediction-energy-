```python
# Cell 0: Imports and settings

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (9,4)
sns.set_theme(style="whitegrid")

# Use your Kaggle path
FILEPATH = "/kaggle/input/carbon-perdiction/Global Energy Review 2025 Dataset.xlsx"
YEARS = [2022, 2023, 2024]

# Cell 1: Detect real header row and read sheets

def detect_header_row(df_no_header: pd.DataFrame, header_key="Publication"):
    """
    Find the row index that contains the real header (looks for 'Publication').
    """
    for i in range(min(30, len(df_no_header))):
        vals = df_no_header.iloc[i, :].astype(str).tolist()
        if any(v == header_key for v in vals):
            return i
    return None

def read_and_clean_sheet(path: str, sheet_name: str) -> pd.DataFrame:
    temp = pd.read_excel(path, sheet_name=sheet_name, header=None, dtype=object)
    h_idx = detect_header_row(temp, header_key="Publication")
    if h_idx is None:
        df = pd.read_excel(path, sheet_name=sheet_name, header=0, dtype=object)
    else:
        df = pd.read_excel(path, sheet_name=sheet_name, header=h_idx, dtype=object)

    # Normalize column names
    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.replace(r"\s+", " ", regex=True)
    )
    # Drop empty cols/rows
    df = df.dropna(axis=1, how="all").dropna(axis=0, how="all").copy()
    df["__Sheet__"] = sheet_name
    return df

world_df = read_and_clean_sheet(FILEPATH, "World")
regions_df = read_and_clean_sheet(FILEPATH, "Regions")

print("World shape:", world_df.shape)
print("Regions shape:", regions_df.shape)
display(world_df.head(5))
display(regions_df.head(5))

# Cell 2: Combine, standardize Region, keep relevant columns

raw = pd.concat([world_df, regions_df], ignore_index=True)

# Standardize Region column
region_col = None
for cand in ["Region", "Regions", "Country/Region", "Area"]:
    if cand in raw.columns:
        region_col = cand
        break
if region_col is None:
    raw["Region"] = "World"
else:
    raw = raw.rename(columns={region_col: "Region"})

# Identify year/meta columns
year_cols = [c for c in ["2022","2023","2024"] if c in raw.columns]
meta_cols = [c for c in ["Publication","Category","Product","Flow","Unit","Region"] if c in raw.columns]

# Keep only the needed columns and rows
df = raw[meta_cols + year_cols].copy()
df = df.dropna(subset=year_cols, how="all")

# Convert year columns to numeric
for c in year_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce")

print("Columns:", df.columns.tolist())
print("Rows:", len(df))
display(df.head(8))

# Cell 3: Melt to long and tag metric groups

long = df.melt(
    id_vars=meta_cols,
    value_vars=year_cols,
    var_name="Year",
    value_name="Value"
).dropna(subset=["Value"]).copy()
long["Year"] = long["Year"].astype(int)

def tag_metric_group(row):
    cat = str(row.get("Category",""))
    flow = str(row.get("Flow",""))
    unit = str(row.get("Unit",""))

    if cat == "Energy" and "Total energy supply" in flow:
        return "Energy supply"
    if cat == "Energy" and "Electricity generation" in flow:
        return "Electricity generation"
    if "CO2" in cat:
        return "CO2"
    if cat == "Technology deployment":
        return "Technology deployment"
    if "Mt CO2" in unit:
        return "CO2"
    if "TWh" in unit:
        return "Electricity generation"
    if unit in {"PJ","EJ"}:
        return "Energy supply"
    return "Other"

long["MetricGroup"] = long.apply(tag_metric_group, axis=1)

print("MetricGroup counts:")
display(long["MetricGroup"].value_counts())
display(long.head(10))

# Cell 4: Pivot blocks to wide and merge

def pivot_block(df_block: pd.DataFrame, value_col="Value"):
    # Ensure Region exists
    if "Region" not in df_block.columns:
        df_block = df_block.assign(Region="World")
    pvt = df_block.pivot_table(
        index=["Year","Region"],
        columns="Product",
        values=value_col,
        aggfunc="first"
    )
    pvt.columns = [str(c) for c in pvt.columns]
    return pvt.reset_index()

energy_supply = pivot_block(long[long["MetricGroup"]=="Energy supply"])
electricity   = pivot_block(long[long["MetricGroup"]=="Electricity generation"])
co2_block     = pivot_block(long[long["MetricGroup"]=="CO2"])

def add_prefix(df, prefix):
    return df.rename(columns={c: f"{prefix}:{c}" for c in df.columns if c not in ["Year","Region"]})

energy_supply = add_prefix(energy_supply, "Energy supply")
electricity   = add_prefix(electricity, "Electricity generation")
co2_block     = add_prefix(co2_block, "CO2")

base = energy_supply.merge(electricity, on=["Year","Region"], how="outer")
base = base.merge(co2_block, on=["Year","Region"], how="outer")

# Unit harmonization: If energy supply looks like PJ (>1000), convert to EJ
E_TOT = "Energy supply:Total"
if E_TOT in base.columns:
    max_val = base[E_TOT].dropna().max()
    if pd.notna(max_val) and max_val > 1000:
        for c in [col for col in base.columns if col.startswith("Energy supply:")]:
            base[c] = base[c] / 1000.0

display(base.head(10))

# Cell 5: Global EDA — trend and intensity

WORLD = base[base["Region"]=="World"].sort_values("Year").set_index("Year")

# Detect CO2 total column
co2_total_candidates = [c for c in base.columns if c.startswith("CO2:") and ("Total" in c or "CO2 emissions" in c)]
CO2_TOTAL = "CO2:Total" if "CO2:Total" in base.columns else (co2_total_candidates[0] if co2_total_candidates else None)

if CO2_TOTAL is None:
    raise ValueError("Could not find CO2 total column. Inspect co2_block columns.")

# Plot total CO2
ax = WORLD[CO2_TOTAL].plot(marker="o", title="Global CO₂ emissions (Mt)")
ax.set_ylabel("Mt CO₂")
plt.show()

# Intensity (Mt/EJ)
if E_TOT in WORLD.columns:
    intensity = (WORLD[CO2_TOTAL] / WORLD[E_TOT]).rename("Intensity (Mt/EJ)")
    ax = intensity.plot(marker="o", title="Global carbon intensity of energy (Mt/EJ)")
    ax.set_ylabel("Mt/EJ")
    plt.show()

# Change 2022 -> 2024
if set(YEARS).issubset(WORLD.index):
    inc_abs = WORLD.loc[2024, CO2_TOTAL] - WORLD.loc[2022, CO2_TOTAL]
    inc_pct = (inc_abs / WORLD.loc[2022, CO2_TOTAL]) * 100
    print(f"Global CO₂ change 2022→2024: {inc_abs:,.0f} Mt ({inc_pct:.2f}%)")

# Cell 6: Regional EDA — who contributed most?

focus_regions = ["United States","European Union","China","India","World"]
panel = base[base["Region"].isin(focus_regions)].copy()
panel = panel.sort_values(["Region","Year"])

if CO2_TOTAL in panel.columns:
    # Line plot by region
    plt.figure(figsize=(10,5))
    for r in focus_regions:
        sub = panel[panel["Region"]==r]
        if not sub.empty:
            plt.plot(sub["Year"], sub[CO2_TOTAL], marker="o", label=r)
    plt.title("CO₂ emissions by region (Mt)")
    plt.ylabel("Mt CO₂"); plt.legend(); plt.show()

    # Compute 2022 -> 2024 change
    def delta_224(g):
        g = g.set_index("Year")
        if set([2022,2024]).issubset(g.index):
            return g.loc[2024, CO2_TOTAL] - g.loc[2022, CO2_TOTAL]
        return np.nan

    changes = (
        panel.groupby("Region")
             .apply(delta_224)
             .dropna()
             .sort_values(ascending=False)
             .rename("Δ CO₂ (2024-2022) Mt")
    )
    print("Change 2022→2024 by region (Mt):")
    display(changes.to_frame())

    # Bar chart of changes (excluding World to focus on countries/blocks)
    ch_plot = changes.drop(labels=[lbl for lbl in ["World"] if lbl in changes.index], errors="ignore")
    ch_plot.plot(kind="bar", title="CO₂ change 2022→2024 by region (Mt)")
    plt.ylabel("Mt CO₂"); plt.show()
else:
    print("CO2 total column not found in panel.")

# Cell 7: Sector (fuel) analysis — which fuels drove changes?

# Detect CO2 fuel columns
CO2_COAL = "CO2:Coal" if "CO2:Coal" in base.columns else None
CO2_OIL  = "CO2:Oil" if "CO2:Oil" in base.columns else None
CO2_GAS  = "CO2:Natural gas" if "CO2:Natural gas" in base.columns else None
CO2_BIO  = "CO2:Bioenergy and waste" if "CO2:Bioenergy and waste" in base.columns else None
CO2_PROC = "CO2:Industrial process" if "CO2:Industrial process" in base.columns else None

fuel_cols = [c for c in [CO2_COAL, CO2_OIL, CO2_GAS, CO2_BIO, CO2_PROC] if c is not None]

if WORLD.index.isin(YEARS).all() and fuel_cols:
    # Plot levels by year
    WORLD[fuel_cols].plot(kind="bar", title="World CO₂ by fuel/process (Mt)")
    plt.ylabel("Mt CO₂"); plt.xticks(rotation=0); plt.show()

    # Contribution to change 2022->2024
    contrib = {}
    for c in fuel_cols:
        contrib[c.replace("CO2:","")] = WORLD.loc[2024, c] - WORLD.loc[2022, c]
    contrib = pd.Series(contrib).sort_values(ascending=False)
    print("Fuel/process contributions to CO₂ change 2022→2024 (Mt):")
    display(contrib.to_frame("Δ Mt"))

    contrib.plot(kind="bar", title="Contribution to CO₂ change (2024-2022) by fuel/process")
    plt.ylabel("Mt CO₂"); plt.show()
else:
    print("Fuel-level CO₂ columns not available or missing years.")

# Cell 8: Emission-factor model (transparent baseline) and predictions with 95% CI

# Detect energy-by-fuel columns
E_COAL = "Energy supply:Coal" if "Energy supply:Coal" in base.columns else None
E_OIL  = "Energy supply:Oil" if "Energy supply:Oil" in base.columns else None
E_GAS  = "Energy supply:Natural gas" if "Energy supply:Natural gas" in base.columns else None

# Basic checks
needed = [CO2_TOTAL, CO2_COAL, CO2_OIL, CO2_GAS, E_COAL, E_OIL, E_GAS]
if not all(col is not None and col in WORLD.columns for col in needed):
    missing = [col for col in ["CO2_TOTAL","CO2_COAL","CO2_OIL","CO2_GAS","E_COAL","E_OIL","E_GAS"] if eval(col) is None or eval(col) not in WORLD.columns]
    raise ValueError(f"Missing required columns for model: {missing}")

yr0 = 2022
ef_coal = WORLD.loc[yr0, CO2_COAL] / WORLD.loc[yr0, E_COAL]
ef_oil  = WORLD.loc[yr0, CO2_OIL]  / WORLD.loc[yr0, E_OIL]
ef_gas  = WORLD.loc[yr0, CO2_GAS]  / WORLD.loc[yr0, E_GAS]

# Baseline = bioenergy + industrial process in 2022 (if available)
baseline = 0.0
if CO2_BIO:  baseline += WORLD.loc[yr0, CO2_BIO]
if CO2_PROC: baseline += WORLD.loc[yr0, CO2_PROC]

def predict_year(y):
    return (
        ef_coal * WORLD.loc[y, E_COAL] +
        ef_oil  * WORLD.loc[y, E_OIL]  +
        ef_gas  * WORLD.loc[y, E_GAS]  +
        baseline
    )

preds = pd.Series({y: predict_year(y) for y in YEARS}, name="Predicted_Mt")
actuals = WORLD.loc[YEARS, CO2_TOTAL].rename("Actual_Mt")
residuals = preds - actuals

# Use 2023–2024 residuals to estimate RMSE -> 95% CI
eval_years = [y for y in YEARS if y != 2022]
rmse = float(np.sqrt(np.mean(residuals.loc[eval_years]**2))) if eval_years else 0.0
ci95_half = 1.96 * rmse

results = pd.DataFrame({
    "Predicted_Mt": preds,
    "Actual_Mt": actuals,
    "Residual_Mt": residuals,
    "CI95_low": preds - ci95_half,
    "CI95_high": preds + ci95_half
})
display(results)

# Plot
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(results.index, results["Actual_Mt"], marker="o", label="Actual")
ax.plot(results.index, results["Predicted_Mt"], marker="s", label="Predicted")
ax.fill_between(results.index, results["CI95_low"], results["CI95_high"], color="tab:blue", alpha=0.15, label="95% CI")
ax.set_title("World CO₂: actual vs predicted")
ax.set_xlabel("Year"); ax.set_ylabel("Mt CO₂"); ax.legend()
plt.show()

print("Calibration (2022 emission factors):")
print(f"- Coal intensity: {ef_coal:.2f} Mt/EJ")
print(f"- Oil intensity:  {ef_oil:.2f} Mt/EJ")
print(f"- Gas intensity:  {ef_gas:.2f} Mt/EJ")
print(f"- Baseline (bio + process): {baseline:,.0f} Mt")
print(f"- RMSE (eval years): {rmse:,.0f} Mt | CI half-width: {ci95_half:,.0f} Mt")

# Cell 9: Clear summary — what increased, who contributed, which sectors mattered

summary_lines = []

# Global change
g_abs = WORLD.loc[2024, CO2_TOTAL] - WORLD.loc[2022, CO2_TOTAL]
g_pct = g_abs / WORLD.loc[2022, CO2_TOTAL] * 100
summary_lines.append(f"Global CO₂ rose by {g_abs:,.0f} Mt from 2022 to 2024 ({g_pct:.2f}%).")

# Regional changes
focus_regions = ["United States","European Union","China","India"]
panel = base[base["Region"].isin(focus_regions)].copy()
panel = panel.sort_values(["Region","Year"])
reg_delta = (panel.pivot(index="Region", columns="Year", values=CO2_TOTAL)
                 .assign(Delta=lambda d: d[2024]-d[2022])
                 .sort_values("Delta", ascending=False))
summary_lines.append("By region (2022→2024 Δ Mt): " + ", ".join([f"{r}: {reg_delta.loc[r,'Delta']:,.0f}" for r in reg_delta.index]))

# Fuel contributions
fuel_cols = [c for c in ["CO2:Coal","CO2:Oil","CO2:Natural gas","CO2:Bioenergy and waste","CO2:Industrial process"] if c in WORLD.columns]
if fuel_cols:
    contrib = {c.replace("CO2:",""): WORLD.loc[2024, c] - WORLD.loc[2022, c] for c in fuel_cols}
    contrib_series = pd.Series(contrib).sort_values(ascending=False)
    top_fuels = ", ".join([f"{k}: {v:,.0f} Mt" for k, v in contrib_series.items()])
    summary_lines.append("Fuel/process contributions (2024–2022): " + top_fuels)

print("\n".join(summary_lines))
```
